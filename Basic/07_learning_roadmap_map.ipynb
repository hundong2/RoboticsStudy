# 07. 전체 학습 로드맵 맵

아래 이미지는 `Basic` 학습 자료 전체를 한눈에 보기 위한 로드맵입니다. 각 단계는 앞의 노트북 순서와 대응됩니다.

![Robotics Basic 전체 학습 로드맵](assets/robotics_learning_roadmap.svg)

## 보는 법

1. `로봇 수학 -> 센서와 인식 -> SLAM -> ROS 2 시스템`은 로봇 시스템의 기본 뼈대입니다.
2. `시뮬레이션`은 실제 장비 없이 이 뼈대를 검증하는 실습 환경입니다.
3. `VLM/VLA`는 perception과 language understanding을 action으로 확장하는 최신 로봇 AI 방향입니다.
4. 논문은 처음부터 깊게 파지 말고, 각 단계의 문제의식이 보일 때마다 연결해서 읽습니다.
