# 05. 시뮬레이션과 Isaac Sim

목표: Isaac Sim이 로봇 개발 흐름에서 어떤 역할을 하는지 이해한다.

## 시뮬레이션을 쓰는 이유

- 실제 로봇 없이 센서/제어/인식 pipeline을 먼저 검증한다.
- 실패해도 장비 손상 위험이 낮다.
- 다양한 환경과 조명 조건을 반복 생성할 수 있다.
- synthetic data를 만들어 인식 모델 학습에 활용할 수 있다.
- RL/imitation learning 실험을 병렬로 돌릴 수 있다.

## Isaac Sim의 위치

Isaac Sim은 NVIDIA Omniverse 기반 로봇 시뮬레이션 플랫폼이다. 물리 시뮬레이션, RTX 기반 센서 렌더링, ROS 2 연동, synthetic data 생성, 로봇 학습 워크플로우에 사용된다.

공식 문서 읽기: [papers/isaac_sim_docs.md](papers/isaac_sim_docs.md)

## 같이 알아야 할 용어

- USD/OpenUSD: 장면, 물체, 재질, 계층 구조를 표현하는 포맷
- URDF: ROS에서 많이 쓰는 로봇 모델 포맷
- sensor simulation: camera, depth, LiDAR, IMU 같은 가상 센서 생성
- domain randomization: 시뮬레이션 환경을 무작위로 바꿔 현실 적응력을 높이는 기법
- sim-to-real: 시뮬레이션에서 만든 모델/정책을 실제 로봇으로 옮기는 과정
- Isaac Lab: Isaac Sim 위에서 로봇 학습 실험을 구성하기 위한 프레임워크

## 입문 실습 순서

1. 예제 로봇과 월드를 실행한다.
2. 카메라/LiDAR 데이터를 ROS 2 topic으로 publish한다.
3. ROS 2에서 RViz로 센서 데이터를 확인한다.
4. 간단한 navigation 또는 manipulation 예제를 실행한다.
5. synthetic dataset을 생성하고 detection/segmentation 학습에 연결한다.
6. Isaac Lab으로 RL 또는 imitation learning 예제를 살펴본다.
