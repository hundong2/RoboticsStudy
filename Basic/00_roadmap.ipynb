# 00. 로봇 기초 로드맵

목표: SLAM, Isaac Sim, VLM, VLA 같은 용어가 어디에 속하는지 먼저 분류하고, 무엇부터 공부할지 결정한다.

## 큰 지도

| 영역 | 핵심 질문 | 대표 키워드 |
|---|---|---|
| 로봇 수학 | 로봇의 위치와 움직임을 어떻게 표현할까? | 좌표계, SE(3), quaternion, Jacobian, 확률 |
| 센서/인식 | 세상을 어떻게 관측할까? | camera, LiDAR, IMU, calibration, detection, segmentation |
| 위치추정/지도 | 내가 어디 있고 주변은 어떻게 생겼을까? | odometry, SLAM, loop closure, occupancy grid |
| 제어/계획 | 목표까지 어떻게 움직일까? | PID, MPC, path planning, trajectory |
| 로봇 미들웨어 | 여러 모듈을 어떻게 연결할까? | ROS 2, topic, service, action, TF, Nav2 |
| 시뮬레이션 | 실제 로봇 없이 어떻게 테스트할까? | Isaac Sim, Gazebo, Isaac Lab, sim-to-real |
| 로봇 학습 | 데이터로 행동을 어떻게 배울까? | imitation learning, reinforcement learning, VLM, VLA, RT-1/RT-2 |

## 추천 진입 순서

1. 좌표계와 센서 데이터의 의미를 익힌다.
2. OpenCV로 이미지 처리와 카메라 모델을 익힌다.
3. SLAM 개념을 공부한다. 처음부터 논문 구현을 목표로 하지 않는다.
4. ROS 2에서 센서, TF, 로봇 상태가 어떻게 흐르는지 본다.
5. Isaac Sim/Gazebo 같은 시뮬레이터에서 센서와 로봇을 띄운다.
6. VLM/VLA는 Transformer와 CLIP 계열 개념을 이해한 뒤 들어간다.

## 용어를 만났을 때 판단법

- SLAM: 위치추정 + 지도작성 문제다. 센서, 확률, 최적화가 핵심이다.
- Isaac Sim: 로봇 시뮬레이션 플랫폼이다. 물리, 센서 렌더링, synthetic data, RL 학습에 쓰인다.
- VLM: Vision-Language Model. 이미지와 텍스트를 함께 이해하는 모델이다.
- VLA: Vision-Language-Action Model. 이미지/언어 입력을 로봇 행동으로 연결하는 모델이다.
- Nav2: ROS 2 기반 자율주행 내비게이션 스택이다.
- TF: 로봇의 여러 좌표계 관계를 시간에 따라 관리하는 시스템이다.
