# 01. 로봇 수학 기초

목표: 로봇 논문과 ROS/SLAM 문서를 읽을 때 막히는 최소 수학을 정리한다.

## 1. 좌표계와 변환

로봇은 항상 여러 좌표계를 가진다.

- `map`: 전역 지도 좌표계
- `odom`: 누적 주행 거리 기준 좌표계
- `base_link`: 로봇 본체 중심
- `camera_link`, `lidar_link`, `imu_link`: 센서 좌표계

핵심은 한 좌표계의 점을 다른 좌표계로 변환하는 것이다. 2D에서는 회전 행렬과 이동 벡터, 3D에서는 homogeneous transform, quaternion, SE(3)를 자주 쓴다.

## 2. 확률 감각

로봇 센서는 항상 불확실하다. 그래서 로봇은 보통 `정답 위치`가 아니라 `위치에 대한 믿음`을 관리한다.

- 평균: 현재 추정값
- 분산/공분산: 얼마나 불확실한지
- 베이즈 필터: 예측과 관측을 반복해 믿음을 갱신하는 틀
- Kalman Filter / EKF / Particle Filter: 대표적인 상태 추정 알고리즘

읽기 자료: [papers/probabilistic_robotics.md](papers/probabilistic_robotics.md)

## 3. 최적화

SLAM, bundle adjustment, calibration은 대부분 오차를 최소화하는 문제로 바뀐다.

- residual: 예측과 관측의 차이
- least squares: 오차 제곱합 최소화
- Jacobian: 파라미터가 바뀔 때 오차가 얼마나 바뀌는지
- graph optimization: pose와 landmark를 노드로 두고 제약을 edge로 둔 최적화

처음에는 수식을 모두 유도하기보다 `무엇을 최소화하는가`를 보는 것이 중요하다.

## 체크리스트

- [ ] 2D 회전 행렬을 설명할 수 있다.
- [ ] homogeneous transform이 왜 필요한지 설명할 수 있다.
- [ ] covariance가 불확실성을 표현한다는 것을 안다.
- [ ] Kalman Filter가 예측-갱신 구조라는 것을 안다.
- [ ] SLAM이 최적화 문제로 표현될 수 있다는 것을 안다.
