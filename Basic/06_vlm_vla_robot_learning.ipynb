# 06. VLM, VLA, 로봇 학습

목표: 최근 로봇 AI 용어를 기초 개념 위에 배치한다.

## VLM과 VLA

- VLM: Vision-Language Model. 이미지와 텍스트를 함께 이해한다. 예: CLIP, BLIP 계열, GPT-4o류 멀티모달 모델.
- VLA: Vision-Language-Action Model. 이미지/언어 입력을 로봇 행동으로 변환한다. 예: RT-2 계열.

VLM은 `무엇을 보고 무엇을 말할 수 있는가`에 가깝고, VLA는 `무엇을 보고 어떤 행동을 할 것인가`에 가깝다.

## 로봇 학습의 대표 방식

| 방식 | 아이디어 | 장점 | 어려움 |
|---|---|---|---|
| imitation learning | 사람이 한 행동을 따라 배움 | 직관적 | 좋은 데모 데이터 필요 |
| reinforcement learning | 보상을 최대화하도록 시행착오 | 복잡한 제어 가능 | 샘플 비용, 보상 설계 |
| offline RL | 수집된 데이터로 정책 학습 | 실제 시행착오 감소 | 데이터 분포 밖 행동 위험 |
| VLA fine-tuning | VLM/LLM 사전학습 지식을 행동에 연결 | 일반화 기대 | 데이터/하드웨어/평가 비용 큼 |

## 현실적인 입문 프로젝트

- 카메라 이미지와 텍스트 프롬프트로 물체 후보를 찾는다.
- 검출 결과를 로봇 좌표계로 변환한다.
- 실제 제어는 처음엔 rule-based로 둔다.
- 이후 imitation learning으로 pick/place 행동을 학습한다.
- 마지막에 VLM/VLA 논문 구조와 비교한다.

핵심은 처음부터 end-to-end VLA를 만들려 하지 않는 것이다. perception, grounding, planning, control을 나누어 디버깅 가능한 시스템을 먼저 만든다.
