# 11. 통합 마무리: MVP 인지/자연어 명령 연동

목표: 지금까지 배운 ROS2, SLAM/Nav2, MoveIt2, 비전, LLM, 모방학습을 하나의 최소 프로젝트로 묶는다.

## MVP 목표 예시

`책상 위 물체를 인식하고, 사용자의 자연어 명령을 받아 로봇이 이동하거나 로봇팔 작업 시퀀스를 실행한다.`

처음 MVP는 완전 자율 로봇이 아니라, 각 모듈이 실제로 연결되는지를 검증하는 수준이면 충분하다.

## MVP 구성도

```text
[User Command]
      |
      v
[LLM/Ollama: task parser]
      |
      +--> [Vision: OpenCV/YOLO] -> [Object pose]
      |
      +--> [Nav2: 이동 목표]
      |
      +--> [MoveIt2: 로봇팔 작업]
      |
      v
[ROS2 launch + logs + safety checks]
```

## 3-4주 이론 수업 기준 묶음

사진 속 커리큘럼처럼 이론 단계에서는 아래 4개를 함께 다루는 것이 좋다.

1. ROS2 기초: Node, Topic, Service, Action, Launch, URDF, TF2
2. SLAM/Nav2: 매핑 원리, 경로 계획, 코스트맵, 행동트리
3. MoveIt2: 모션 플래닝, Pick & Place, 작업 시퀀스
4. OMX 활용 모방학습: 텔레옵, 데이터 수집, 학습, 추론 개념

그 다음 OpenCV/YOLO와 LLM/Ollama를 붙여 인지/자연어 명령 MVP로 확장한다.

## MVP 단계별 구현 순서

1. ROS2 launch로 모든 노드 실행 구조를 만든다.
2. 카메라 topic을 받아 OpenCV/YOLO로 물체를 탐지한다.
3. 탐지 결과를 TF 기준 좌표로 변환한다.
4. 이동이 필요한 작업은 Nav2 goal로 보낸다.
5. 조작이 필요한 작업은 MoveIt2 pick/place sequence로 보낸다.
6. 자연어 명령은 LLM/Ollama가 제한된 action list 중 하나로 변환한다.
7. 실행 결과와 실패 원인을 bag/log로 저장한다.
8. 반복 작업은 텔레옵 데이터로 수집해 모방학습 실험으로 확장한다.

## 완료 기준

- [ ] 자연어 명령이 구조화된 action sequence로 변환된다.
- [ ] 비전 모듈이 물체 후보와 위치를 출력한다.
- [ ] Nav2 또는 MoveIt2 중 하나 이상과 실제로 연결된다.
- [ ] 실패 시 멈추고 로그를 남긴다.
- [ ] 다음 단계로 모방학습 데이터 수집 계획을 세울 수 있다.
