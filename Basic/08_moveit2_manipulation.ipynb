# 08. MoveIt2와 로봇팔 조작

목표: 모바일 로봇 주행과 별개로, 로봇팔이 물체를 집고 옮기는 manipulation pipeline을 이해한다.

## MoveIt2가 맡는 역할

MoveIt2는 ROS 2 기반 로봇팔 motion planning 프레임워크다. 로봇팔의 관절 제한, 충돌 가능성, 목표 pose를 고려해서 실행 가능한 움직임을 계산한다.

- motion planning: 시작 관절 상태에서 목표 pose까지의 경로 계산
- collision checking: 로봇 자신과 주변 물체의 충돌 검사
- inverse kinematics: end-effector 목표 pose를 관절 각도로 변환
- trajectory execution: 계산된 경로를 controller로 전달
- planning scene: 주변 물체와 작업 환경 상태 관리

## Pick & Place 기본 흐름

```text
물체 인식 -> 물체 pose 추정 -> 접근 pose 생성 -> grasp -> lift -> place pose 이동 -> release
```

처음에는 딥러닝 grasping을 바로 하지 말고, 물체 위치를 알고 있다고 가정한 rule-based pick & place부터 실습하는 것이 좋다.

## 작업 시퀀스 관점

MoveIt2 자체는 '무엇을 먼저 하고 다음에 무엇을 할지'를 모두 결정하는 도구가 아니다. 보통 아래 계층으로 나눈다.

| 계층 | 예시 |
|---|---|
| 명령 이해 | '빨간 컵을 집어 오른쪽에 놓아줘' |
| 작업 계획 | cup 탐색 -> 접근 -> 집기 -> 이동 -> 놓기 |
| 모션 계획 | 각 단계의 관절 궤적 계산 |
| 제어 실행 | controller가 모터 명령으로 실행 |

LLM은 주로 명령 이해와 작업 계획에 붙고, MoveIt2는 모션 계획과 실행에 붙는다.

## 학습 체크리스트

- [ ] URDF/SRDF가 무엇인지 안다.
- [ ] end-effector, joint state, planning scene을 설명할 수 있다.
- [ ] inverse kinematics와 motion planning의 차이를 안다.
- [ ] Pick & Place를 여러 단계의 작업 시퀀스로 나눌 수 있다.
- [ ] MoveIt2가 SLAM/Nav2와 다른 문제를 푼다는 것을 안다.
