# 02. 센서, 인식, 컴퓨터 비전

목표: 로봇이 세상을 관측하는 방식과 OpenCV/딥러닝 인식의 위치를 이해한다.

## 주요 센서

| 센서 | 얻는 데이터 | 장점 | 주의점 |
|---|---|---|---|
| RGB camera | 이미지 | 저렴하고 정보량이 큼 | 조명, 모션 블러, 깊이 부재 |
| Depth camera | RGB-D | 가까운 거리 3D 정보 | 실외/반사체에 취약 |
| LiDAR | point cloud | 거리 정확도, 야외 강함 | 비용, sparse data |
| IMU | 가속도/각속도 | 빠른 motion 추정 | drift 누적 |
| wheel encoder | 바퀴 회전 | 주행 거리 추정 | 미끄러짐에 취약 |

## Camera pipeline

1. 이미지 획득
2. 왜곡 보정 및 calibration
3. feature 추출 또는 딥러닝 추론
4. 추적, 검출, segmentation, depth 추정
5. 로봇 좌표계로 변환

OpenCV는 1-4단계의 전통적 비전 알고리즘을 익히기에 좋고, PyTorch/ONNX Runtime은 딥러닝 모델 배포에 자주 쓰인다.

## Feature 기반 vs 딥러닝 기반

- Feature 기반: ORB, SIFT, FAST처럼 이미지의 점/패턴을 찾는다. SLAM에서 오래 쓰였고 빠르다.
- 딥러닝 기반: detector, segmenter, depth estimator, VLM으로 의미 정보를 얻는다.
- 실무에서는 둘 중 하나만 쓰기보다, geometric pipeline과 semantic perception을 함께 붙이는 경우가 많다.
