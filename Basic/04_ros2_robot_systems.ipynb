# 04. ROS 2와 로봇 시스템 구조

목표: 알고리즘을 실제 로봇 시스템에 붙일 때 필요한 연결 방식을 이해한다.

## ROS 2 핵심 개념

- node: 기능 단위 프로세스
- topic: 지속적으로 흐르는 데이터 채널
- service: 요청-응답 RPC
- action: 오래 걸리는 목표 수행
- message: topic/service/action의 데이터 형식
- TF: 시간에 따른 좌표계 관계
- TF2: ROS 2에서 좌표계 변환을 시간과 함께 관리하는 라이브러리
- URDF: 로봇 링크와 관절 구조를 기술하는 모델 파일
- launch: 여러 노드와 파라미터를 한 번에 실행하는 구성
- bag: 센서/토픽 데이터 기록 파일

## 전형적인 모바일 로봇 구성

```text
camera/lidar/imu -> sensor drivers -> perception/slam -> map/pose
map/pose/goal -> planner -> controller -> motor driver -> robot
                         ^
                         |
                    diagnostics/logging
```

실무에서는 알고리즘 성능만큼 시간 동기화, 좌표계, 로그, 재현성이 중요하다.

## Nav2가 하는 일

Nav2는 ROS 2 기반 navigation stack이다.

- map server
- localization
- global planner
- local controller
- costmap
- behavior tree 기반 task orchestration

SLAM으로 만든 지도를 Nav2에 넣고, 목표 지점까지 이동시키는 흐름이 모바일 로봇의 대표 실습이다.
