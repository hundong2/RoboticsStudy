# 10. 비전, YOLO, LLM/Ollama 연동

목표: 로봇에서 OpenCV/YOLO 같은 비전 모듈과 LLM을 어디에 붙이는지 이해한다.

## 비전 모듈의 역할

- OpenCV: 카메라 입력, 전처리, 색/형태 기반 탐지, calibration, geometry 처리
- YOLO: 물체 bounding box 탐지
- Segmentation: 픽셀 단위 물체 영역 추정
- Pose estimation: 물체 또는 사람의 자세 추정
- Depth/3D 변환: 이미지 좌표를 로봇 좌표계로 연결

비전 모듈은 로봇에게 `무엇이 어디에 있는가`를 알려주는 계층이다.

## LLM/Ollama의 역할

Ollama는 로컬에서 LLM을 실행하기 위한 도구로 볼 수 있다. 로봇에서 LLM은 보통 저수준 제어가 아니라 자연어 명령 해석과 작업 분해에 사용한다.

예시:

```text
사용자: 책상 위 빨간 컵을 왼쪽 상자에 넣어줘
LLM: detect(red cup) -> estimate_pose(cup) -> pick(cup) -> place(left_box)
로봇 시스템: YOLO/OpenCV + TF + MoveIt2/Nav2가 실제 실행
```

## LLM을 붙일 때 주의할 점

- LLM 출력은 바로 모터 명령으로 쓰지 않는다.
- 허용된 tool/action 목록 안에서만 선택하게 한다.
- 실행 전 perception 결과와 로봇 상태를 검증한다.
- 실패하면 멈추고 재계획하도록 만든다.
- 로그를 남겨 어떤 명령이 어떤 action sequence로 바뀌었는지 추적한다.

## 통합 구조

```text
자연어 명령
  -> LLM/Ollama: 작업 분해
  -> Vision: 물체 탐지/위치 추정
  -> TF: 로봇 좌표계 변환
  -> Nav2 또는 MoveIt2: 주행/조작 계획
  -> Controller: 실행
  -> 상태 확인 및 재계획
```

## 체크리스트

- [ ] OpenCV와 YOLO의 역할 차이를 설명할 수 있다.
- [ ] LLM을 low-level controller로 쓰면 위험한 이유를 안다.
- [ ] 자연어 명령을 action sequence로 바꾸는 구조를 설계할 수 있다.
- [ ] perception 결과를 TF/MoveIt2/Nav2에 연결해야 한다는 것을 안다.
