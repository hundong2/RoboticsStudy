# 09. OMX 활용 모방학습 개념

목표: OpenMANIPULATOR-X 같은 교육용 로봇팔을 기준으로 텔레옵, 데이터 수집, 학습, 추론 흐름을 이해한다.

## OMX란?

여기서 OMX는 보통 ROBOTIS OpenMANIPULATOR-X 계열 로봇팔을 의미하는 맥락으로 다룬다. ROS/MoveIt2 실습, Pick & Place, 모방학습 데이터 수집에 자주 쓰이는 소형 로봇팔이다.

중요한 점은 특정 장비명이 아니라 `로봇팔 + 카메라 + 텔레옵 + 데이터 수집 + 정책 학습`으로 이어지는 실습 구조다.

## 모방학습 pipeline

```text
텔레옵 조작 -> 관측/행동 데이터 저장 -> 데이터 정리 -> 정책 학습 -> 추론 실행 -> 실패 케이스 재수집
```

- 텔레옵: 사람이 조이스틱, 키보드, 마스터 장치로 로봇을 조작한다.
- 데이터: 이미지, joint state, gripper state, action, timestamp를 함께 저장한다.
- 학습: 관측을 입력으로 받고 다음 action을 예측하는 정책을 학습한다.
- 추론: 새 관측에서 action을 내고 실제 로봇 또는 시뮬레이터에서 실행한다.

## 데이터 설계가 핵심

모방학습은 모델보다 데이터 품질이 먼저다.

| 항목 | 기록 이유 |
|---|---|
| RGB image | 물체와 작업 상태 인식 |
| depth 또는 pose | 물체까지의 거리와 위치 추정 |
| joint state | 현재 로봇팔 자세 |
| gripper state | 집기/놓기 상태 |
| action | 모델이 따라 배울 목표 |
| task label | 어떤 작업 데이터인지 구분 |

## VLA와의 연결

RT-1/RT-2 같은 VLA 논문도 결국 큰 관점에서는 `관측 + 언어 지시 -> 행동`을 학습한다. OMX 실습은 그 축소판이다.

처음에는 거대한 VLA를 직접 만들기보다, 한 가지 작업에 대해 imitation learning dataset을 만들고 작은 policy를 학습하는 방식이 현실적이다.

## 체크리스트

- [ ] 텔레옵 데이터와 자율 실행 데이터의 차이를 안다.
- [ ] observation과 action을 명확히 정의할 수 있다.
- [ ] timestamp 동기화가 왜 중요한지 안다.
- [ ] 실패 데이터를 다시 수집해 dataset을 개선하는 루프를 설명할 수 있다.
