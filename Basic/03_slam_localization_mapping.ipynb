# 03. SLAM, 위치추정, 지도작성

목표: SLAM을 하나의 마법 같은 알고리즘이 아니라 여러 하위 문제의 조합으로 이해한다.

## SLAM이란?

SLAM은 Simultaneous Localization and Mapping의 약자다. 로봇이 미지의 환경에서 움직이면서 동시에 자신의 위치와 지도를 추정하는 문제다.

- localization: 내가 어디 있는가
- mapping: 주변 환경은 어떻게 생겼는가
- loop closure: 이전에 왔던 장소를 다시 알아보는가
- relocalization: 위치를 잃었을 때 다시 찾을 수 있는가

## SLAM 종류

| 종류 | 입력 | 대표 특징 |
|---|---|---|
| Visual SLAM | camera | 저렴하지만 조명/텍스처 영향 큼 |
| RGB-D SLAM | RGB-D camera | 실내에서 직관적 |
| LiDAR SLAM | LiDAR | 거리 정확도와 강건성이 좋음 |
| Visual-Inertial SLAM | camera + IMU | 빠른 움직임에 유리 |
| Semantic SLAM | sensor + recognition | 지도에 의미 정보를 추가 |

## 기본 pipeline

1. 센서 데이터 수신
2. feature/scan matching
3. motion estimation
4. local map update
5. loop closure detection
6. pose graph optimization
7. map export 또는 navigation stack 연결

## 실무 진입 과제

- ROS 2 bag 데이터로 SLAM 패키지를 실행해 본다.
- 결과로 나오는 trajectory와 map을 확인한다.
- `map -> odom -> base_link -> sensor` TF 관계를 시각화한다.
- 실패 케이스를 기록한다: 조명, 빠른 회전, feature 부족, 동적 객체, 센서 시간 동기화 문제.

## Nav2와 연결되는 개념

SLAM으로 지도를 만들었다면 다음 단계는 Nav2로 그 지도 위에서 이동하는 것이다.

- costmap: 장애물, inflation, unknown space를 비용으로 표현한 지도
- global planner: 현재 위치에서 목표까지 큰 경로를 계산
- local planner/controller: 근처 장애물과 로봇 속도를 고려해 실제 주행 명령 생성
- behavior tree: navigation 작업을 조건과 행동의 트리로 구성
- recovery behavior: 경로 실패, 장애물, localization 불안정 상황에서 복구 행동 수행

실무에서는 `지도 생성`보다 `지도 위에서 안정적으로 움직이는 것`이 더 오래 걸리는 경우가 많다.
